```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.displot(titanic['fare'],bins=10,kde=True,rug=True);
```

```python
sns.jointplot(x='age',y='fare',data=titanic,kind='hex');
```

```python
sns.rugplot(titanic['fare']);
```

```python
sns.kdeplot(titanic['fare']);
```

```python
sns.relplot(x=titanic['age'],y=titanic['fare'],data=titanic,size='pclass',sizes=(10,100),hue='pclass',col='class',col_wrap=2);
```

```python
sns.relplot(x='age',y='fare',data=titanic,kind='line',estimator=None);
```

```python
sns.barplot(x='sex',y='fare',data=titanic,estimator = np.std,order=['female','male'] );
```

```python
sns.countplot(x='sex',data=titanic,order=['female','male']);
```

```python
sns.violinplot(x='pclass',y='fare',data=titanic,hue='sex',split=True);
```

```python
sns.stripplot(x='pclass',y='fare',hue='sex',data=titanic,jitter=False,dodge=True);
```

```python
sns.swarmplot(x='pclass',y='fare',data=titanic);
```

```python
sns.catplot(x='pclass',y='fare',data=titanic,hue='sex',kind='boxen',dodge=True);
```

```python
t = titanic.corr()
```
